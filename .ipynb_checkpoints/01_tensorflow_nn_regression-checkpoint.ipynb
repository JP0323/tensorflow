# Neural Network Regression with TensorFlow

1. **Create a Model**
2. **Compile a Model**
3. **Fit the Model**
4. **Evaluate the Model**

```python
# Required Packages
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
print(tf.__version__)
```

```python
#Example 
# Make a dataset
X = np.arange(-100, 100, 4)
X
```

```python
# Make labels for the dataset
y = np.arange(-90, 110, 4)
y
```

```python
#logic/comparision
y = X + 10
y
```

## Split data into training/test set

```python
# Split data into train and test sets
X_train = X[:40] # first 40 examples (80% of data)
y_train = y[:40]

X_test = X[40:] # last 10 examples (20% of data)
y_test = y[40:]

len(X_train), len(X_test)
```

## Visualizing the data

```python
plt.figure(figsize=(10, 7))
# Plot training data in blue
plt.scatter(X_train, y_train, c='b', label='Training data')
# Plot test data in green
plt.scatter(X_test, y_test, c='g', label='Testing data')
# Show the legend
plt.legend();
```

## Building the model

```python
# Set random seed
tf.random.set_seed(42)

# Create a model (same as above)
model = tf.keras.Sequential([
  tf.keras.layers.Dense(1, input_shape=[1]) # define the input_shape to our model
])

# Compile model (same as above)
model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])
```

```python
model.summary()
```

```python
# Fit the model to the training data
model.fit(X_train, y_train, epochs=100) #use verbose to hide the data
```

```python
# Make predictions
y_preds = model.predict(X_test)
```

```python
# View the predictions
y_preds
```

```python
def plot_predictions(train_data=X_train, 
                     train_labels=y_train, 
                     test_data=X_test, 
                     test_labels=y_test, 
                     predictions=y_preds):
  """
  Plots training data, test data and compares predictions.
  """
  plt.figure(figsize=(10, 7))
  # Plot training data in blue
  plt.scatter(train_data, train_labels, c="b", label="Training data")
  # Plot test data in green
  plt.scatter(test_data, test_labels, c="g", label="Testing data")
  # Plot the predictions in red (predictions were made on the test data)
  plt.scatter(test_data, predictions, c="r", label="Predictions")
  # Show the legend
  plt.legend();
```

```python
plot_predictions(train_data=X_train,
                 train_labels=y_train,
                 test_data=X_test,
                 test_labels=y_test,
                 predictions=y_preds)
```

```python
# Evaluate the model on the test set
model.evaluate(X_test, y_test)
```

```python
# Calculate the mean absolute error
mae = tf.metrics.mean_absolute_error(y_true=y_test, 
                                     y_pred=y_preds)
mae #this is wrong we need to squeeze to the same shape 
```

```python
def mae(y_test, y_pred):
  """
  Calculuates mean absolute error between y_test and y_preds.
  """
  return tf.metrics.mean_absolute_error(y_test,
                                        y_pred)
  
def mse(y_test, y_pred):
  """
  Calculates mean squared error between y_test and y_preds.
  """
  return tf.metrics.mean_squared_error(y_test,
                                       y_pred)
```

## Building a better model by changing different values

1. Increase the number of layers
2. Change the optimizer
3. Increase the number of epochs

```python
model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(10, input_shape=[1]),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model_1.compile(loss=tf.keras.losses.mae,
               optimizer=tf.keras.optimizers.Adam(),
               metrics=['mae'])

model_1.fit(X_train, y_train, epochs=500)
```

```python
y_preds = model_1.predict(X_test)
X_test, y_test, y_preds
```

```python
plot_predictions(train_data=X_train,
                 train_labels=y_train,
                 test_data=X_test,
                 test_labels=y_test,
                 predictions=y_preds)
```

```python
# Save a model using the SavedModel format
model_1.save('best_model_SavedModel_format')
```

```python
# Save a model using the HDF5 format
model_1.save("best_model_HDF5_format.h5") # note the addition of '.h5' on the end
```

```python
# Load a model from the SavedModel format
loaded_saved_model = tf.keras.models.load_model("best_model_SavedModel_format")
loaded_saved_model.summary()
```
